```python
import numpy as np
import pandas as pd
import os
from PIL import Image
```

```python
def export_imagenet(save_path, train_csv_path, test_csv_path, val_pct=0.2, file_format=".png"):
    
    def create_directories():
        try:
            os.mkdir(save_path)
        except OSError:
            print ("Directory %s not created" % save_path)

        for set_type in ["train", "valid", "test"]:
            try:
                os.mkdir(save_path+"\\"+set_type)
                for i in range(10):
                    os.mkdir(save_path+"\\"+set_type+"\\"+str(i))
            except OSError:
                print ("Directory not created")
    
    
    # Read train and test dataset, split train dataset
    df_train = pd.read_csv(train_csv_path)
    df_validation = df_train.sample(frac=val_pct)
    df_train = df_train.drop(df_validation.index)
    df_test = pd.read_csv(test_csv_path)
    
    create_directories()

    def export_csv(set_type, df):
        for index, row in df.iterrows():
            label = row['label']
            image = row.drop('label').values.reshape(28, 28)

            img = Image.fromarray(np.uint8(image) , 'L')

            image_path = save_path+"\\"+set_type+"\\"+str(label)+"\\"

            img.save(image_path+str(index)+file_format)
            
            if index%1000==0:
                print(index, set_type)
                    
    export_csv("train", df_train)
    export_csv("valid", df_validation)
    export_csv("test", df_test)
    
```

```python
export_imagenet("myimagenet","data/fashion-mnist_train.csv", "data/fashion-mnist_test.csv")
```
